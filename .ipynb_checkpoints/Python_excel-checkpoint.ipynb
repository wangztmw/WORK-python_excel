#### 操作财务数据

```python
import pandas as pd

# 1. 读取Excel文件
df_income = pd.read_excel("新能源汽车财务报表.xlsx", sheet_name="利润表 (单位：万元)")
df_balance = pd.read_excel("新能源汽车财务报表.xlsx", sheet_name="资产负债表 (单位：万元)")
df_cash = pd.read_excel("新能源汽车财务报表.xlsx", sheet_name="现金流量表 (单位：万元)")

# 2. 数据分析示例
# 计算毛利率
revenue = df_income[df_income['项目'] == '营业收入'].iloc[0, 1]
gross_profit = df_income[df_income['项目'] == '毛利'].iloc[0, 1]
gross_margin = gross_profit / revenue
print(f"毛利率：{gross_margin:.2%}")

# 3. 添加财务指标
current_assets = df_balance[df_balance['资产'] == '流动资产合计'].iloc[0, 1]
current_liabilities = df_balance[df_balance['负债和所有者权益'] == '流动负债合计'].iloc[0, 3]
current_ratio = current_assets / current_liabilities
print(f"流动比率：{current_ratio:.2f}")

# 4. 可视化示例 (需安装matplotlib)
import matplotlib.pyplot as plt
expenses = df_income.iloc[3:7, [0, 1]]
expenses.plot(kind='bar', x='项目', title='期间费用分析')
plt.ylabel('万元')
plt.savefig('费用分析.png')

# 5. 导出到新Excel
with pd.ExcelWriter('财务分析报告.xlsx') as writer:
    df_income.to_excel(writer, sheet_name='利润表', index=False)
    
    # 添加分析结果
    analysis_df = pd.DataFrame({
        '指标': ['毛利率', '流动比率'],
        '值': [gross_margin, current_ratio]
    })
    analysis_df.to_excel(writer, sheet_name='财务指标', index=False)
```
